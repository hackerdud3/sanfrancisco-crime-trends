## Project Proposal Summary:  

The purpose of this project is to create an interactive and comprehensive data visualization tool,
specifically focused on property crimes in San Francisco, including theft, burglary, vandalism,
and vehicle-related incidents. This system will provide detailed information to users about
property crimes in San Francisco. The visual details allow users to explore and examine these
patterns to identify property crime hotspots, to understand crime trends in certain neighborhoods,
and to analyze time trends in property crime rates. This visualization tool's main purpose is to
improve public safety awareness, to help law enforcement allocate resources effectively, and to
assist city officials in addressing property crime through a user-friendly interface.

## Questions:  

* What are the most frequently reported property crimes in San Francisco?
* What are the property crime rates in various districts around San Francisco?
* What is the property crime trend over a period of time, either monthly or yearly?
* Are there specific times where property crimes are more active in San Francisco?
* XWhat is the average time gap between the incident date and reported date for property
crimes?



## Target Users.

* City Officials - City officials would use the visualization of property crimes in San
Francisco to allocate resources to law enforcement and to develop better crime prevention
strategies.
* Law Enforcement - Law enforcement would use the visualization of property crimes in
San Francisco to identify districts with high levels of property crime and deploy more
police patrols to those areas.
* Citizens - Citizens would use the visualization of property crimes in San Francisco to
stay informed about the property crime trend in their districts and take precautions to
protect their property.

## End Goals:  
* Develop a clear, visual tool for exploring property crime data in San
Francisco.
* Help users identify trends and patterns in property crime over time.
* Enable better decision-making for city officials and law enforcement in resource
allocation and crime prevention.
* Improve public awareness and understanding of the property crime landscape in their
city.
* Understand patterns of property crime for safer community planning and infrastructure
development.

```python
# clean headings
def clean_headers(df):
  import re

  # create a list of column headers
  df_col_list = list(df)

  for i,x in enumerate(df_col_list):

    # if column header starts with a digit, add an underscore to beginning
    if x[0].isdigit():
      x = '_'+x

    # use list comprehensions to lowercase characters
    # and replace spaces with underscores
    x = x.lower().replace(' ','_')
    # use regex to replace non-alpha-numerics with underscores
    x = re.sub('[^0-9a-z]+', '_', x)
    df_col_list[i] = x

  # combined list comprehension and regex
  # df_col_list = [re.sub('[^0-9a-z]+', '_', x).lower().replace(" ", "_") for x in df_col_list]

  # assign cleaned headers directly
  df.columns = df_col_list
```

```python
# initial EDA function
def initial_eda(df):

  # shape of the data
  print('shape:',df.shape)
  print('\n-------------\n')

  # check how many records are in the dataset and if there any NA
  print('info:')
  print(df.info())
  print('\n-------------\n')

  # get statistics on data frame
  print('describe:')
  print(df.describe())
  print('\n-------------\n')

  # find NA values, sorted by columns with most NA entries
  print('Sort by most NA:')
  print(df.isnull().sum(0).sort_values(ascending=False).head(17))
  print('\n-------------\n')

  # identify missing data with a heat map (white color are nan's)
  # import matplotlib.pyplot as plt
  # import seaborn as sns
  print('Heat Map:')
  plt.figure(figsize=(24,8))
  sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
  plt.show()
  print('\n-------------\n')

  df_col_list = list(df)
  for x in df_col_list:
    print(f'{x}: {df[x].unique().size}\n\n{df[x].value_counts()}\n\n',df[x].unique())
    print('\n-------------\n')
```
